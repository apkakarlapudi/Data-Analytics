```python
leaf_data = pd.read_csv("leaf.csv")
leaf_data.head()
```

```python
# Q1. Eccentricity of the leaf is a measure of how much the shape 
# of the leaf varies from a perfect circle. Unfortunately the dataset 
# is missing values in the eccentricity column. How many missing values 
# are there in this column?
```

```python
leaf_data.isna().sum()
```

```python
# Ans Q1: There are five missing values in the Eccentricity column. 
```

```python
# Q2. Fill in these missing values with something reasonable. 
# Show the head of the dataframe.
```

```python
ld = leaf_data
ecc_avg = ld['eccentricity'].mean()
print(ecc_avg) 
print()
ld['eccentricity'].fillna(ecc_avg, inplace = True)
ld.head()
```

```python
#Q2: I replaced the NaN values with the average eccentricity value. 

# Q3. Standardize the eccentricity column. 
# Where value_std = (value - mean(value)) / std(value). 
# Display the head of the new data frame.
```

```python
norm_ecc = (ld['eccentricity'] - ld['eccentricity'].mean())/ld['eccentricity'].std()
ld['eccentricity'] = norm_ecc
ld.head()
```

```python
#Q4. Plot a scatter plot between smoothness and standardized eccentricity. 
# Place smoothness on the X axis.
ld.plot(x = 'smoothness', y = 'eccentricity', style = 'o')
```

```python
# Q5 
ld.plot.scatter('smoothness', 'eccentricity', c = 'class', colormap='jet')

'''
We can see from this color coded graph that:
- The overall trend of the graph is negatively correlated (slight neg. slope)
- Class 4 is strongly clustered together by eccentricity value
- Class 2 (light blue) and class 3 (yellow) are almost not clustered amongst themselves
- Class 1 is mildly clustered together 
- Classes 1 and 2 appear clustered about an eccentricity value between -.5 and -1

Ultimately, we can infer that: 
- class 4 appears to be a unique group
- classes 1 and 2 appear relatead
- class 3 appears mildly related to classes 1 and 2

'''
```

```python
# Q6 
ld['eccentricity'].corr(ld['smoothness'])
```
